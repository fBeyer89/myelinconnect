```python
v, f, d = read_vtk('/scr/ilz3/myelinconnect/all_data_on_simple_surf/surfs/lowres_rh_d.vtk')
```

```python
face_set= f[0:500]
vertex_set = v[np.unique(face_set.flatten())]
idx = np.unique(face_set.flatten())
face_set_reidx = np.zeros_like(face_set)
for i in range(face_set.shape[0]):
    for j in range(face_set.shape[1]):
        face_set_reidx[i,j] = np.where(idx==face_set[i,j])[0][0]
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(azim=90, elev=90)
ax.set_axis_off()
p3dcollec = ax.plot_trisurf(vertex_set[:, 0], vertex_set[:, 1], vertex_set[:, 2],triangles=face_set_reidx, linewidth=1,antialiased=True,
                                color='white', shade=False)
```
