```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from vtk_rw import read_vtk, write_vtk
from plotting import plot_surf_stat_map, crop_img
```

```python
def brain_fig(plot_list):
    sns.set_style('white')
    n = len(plot_list)
    rows = int(n/2)
    fig = plt.figure(figsize=(12,rows*5))
    for img in range(n):
        ax = fig.add_subplot(rows,2,img+1)
        plt.imshow(plot_list[img])
        ax.set_axis_off()
    fig.tight_layout()
    return fig
```

```python
mesh_dir = '/scr/ilz3/myelinconnect/new_groupavg/surfs/lowres/inflated/'
```

```python
cropped_img = []
```

```python
infl=10
rh_sulc=np.load(mesh_dir+'rh_lowres_new_taubin_%i_sulc.npy'%infl)
rv, rf, _ = read_vtk(mesh_dir+'rh_lowres_new_taubin_%i.vtk'%infl)
lh_sulc=np.load(mesh_dir+'lh_lowres_new_taubin_%i_sulc.npy'%infl)
lv, lf, ld = read_vtk(mesh_dir+'lh_lowres_new_taubin_%i.vtk'%infl)

sns.set_style('white')
for (elev, azim) in [(180, 0), (180, 180)]:
    plot=plot_surf_stat_map(lv, lf, bg_map=lh_sulc, bg_on_stat=True,
                            darkness=0.5, elev=elev,azim=azim, figsize=(10,8.5),threshold=1e-50)
    cropped_img.append(crop_img(plot))
    
for (elev, azim) in [(180, 0), (180, 180)]:
    plot=plot_surf_stat_map(rv, rf, bg_map=rh_sulc, bg_on_stat=True,
                            darkness=0.5, elev=elev,azim=azim, figsize=(10,10),threshold=1e-50)
    cropped_img.append(crop_img(plot))
```
